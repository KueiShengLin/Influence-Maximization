```python
%matplotlib inline
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from diffusion_model import IndependentCascade
```

```python
GRAPH_DATA = pd.read_csv("graph30.txt", sep=" ", header=None)
GRAPH_DATA.columns = ['node_1', 'node_2']
G = nx.from_pandas_edgelist(GRAPH_DATA, target='node_1', source='node_2')
IC = IndependentCascade(G, 0.1)
```

```python
# graph
pos = nx.shell_layout(G)
nx.draw(G, pos=pos, node_size=30)
plt.show()
```

```python
# RanCas(s) is submodularity
submodularity = [1 - (1 - 0.5) ** l for l in range(20)]
x = [l for l in range(20)]
plt.plot(x, submodularity)
plt.show()
```

```python
# activate node 
ith_activate_nodes = IC.run([7, 14])
pos = nx.shell_layout(G)    
nx.draw(G,pos, node_size=30)
nx.draw_networkx_nodes(ith_activate_nodes, pos=pos, node_color='blue', node_size=30)
plt.show()
```
